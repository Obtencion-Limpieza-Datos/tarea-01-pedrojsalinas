```python
import ipywidgets as widgets
import pandas as pd
```

```python
data = pd.read_csv("data/data-educacion-ecuador.csv")
#print data.keys()
cantones=data.Canton.unique()
datos=[]
for canton in cantones:
    datos.append(u"%s"%canton.decode("utf-8"))
```

```python
w = widgets.Dropdown(
    options=datos,
    value='CUENCA',
    description='Canton:',
    disabled=False,
)
display(w)
```

```python
data[data['Canton']==w.value]
```
